```python
import numpy as np
import scipy.integrate as sci
import matplotlib.pyplot as plt
```

$$ mx'' + bx' + kx = f $$

```python
class Parameters(object):
    pass
param = Parameters()
param.m = 1
param.b = 10
param.k = 20
param.p_coef = 100
param.i_coef = 20
param.d_coef =20
param.x_fin = 1

sim_param = Parameters()
sim_param.t0 = 0
sim_param.tf = 20
sim_param.dt = 0.01
```

```python
class harm:
    def __init__ (self, p, sim_param):
        self.p = param
        self.sim_param = sim_param
    def proportional_ode(self, t, x):
        r = x[0]
        v = x[1]
        f = self.p.p_coef*(self.p.x_fin - r)
        dxdt = np.zeros(2)
        dxdt[0] = x[1]
        dxdt[1] = 1/self.p.m * (f - self.p.b - self.p.k*r)
        
        return dxdt
    def proportional(self, x0=[0, 0]):
        t_sol = np.arange(self.sim_param.t0, self.sim_param.tf, self.sim_param.dt)
        sol = sci.solve_ivp(self.proportional_ode, (self.sim_param.t0, self.sim_param.tf), x0, t_eval = t_sol)
        x_sol = sol.y.T
        fig1, ax1 = plt.subplots(1)
        ax1.plot( sol.t, x_sol[:, 0], label='x(t)', color='r')
        #return x_sol
        
```
